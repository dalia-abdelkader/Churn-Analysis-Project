### the model predicts well on the class non-churn but can't predict the churn class because the data is imbalanced



```python
from sklearn.utils import resample
df_majority = df_new2[df_new2['churn']==0]
df_minority = df_new2[df_new2['churn']==1]
# upsample minority class
df_minority_upsampled = resample(df_minority,
                                 replace=True,    # sample with replacement
                                 n_samples= 8048, # to match majority class
                                 random_state=42)
# downsample majority class
df_majority_downsampled = resample(df_majority,
                                 replace=True,    # sample with replacement
                                 n_samples= 8048, # to match majority class
                                 random_state=42)
df_up_down_sampled  = pd.concat([df_minority_upsampled, df_majority_downsampled])
```

```python
X = df_new2.drop(columns=['id','churn'])
y = df_up_down_sampled['churn']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
